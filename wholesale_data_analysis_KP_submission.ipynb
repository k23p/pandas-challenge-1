## Part 2: Transform the Data
Do we know that this client spent the more money than client 66037? If not, how would we find out? Transform the data using the steps below to prepare it for analysis.

```python
# Create a new column 'subtotal' by multiplying 'unit_price' and 'qty'
df['subtotal'] = df['unit_price'] * df['qty']
df[['unit_price', 'qty', 'subtotal']].head()
```

```python
# Create a column for shipping price.
# Assume a shipping price of $7 per pound for orders over 50 pounds and $10 per pound for items 50 pounds or under.


def shipping_calculate(total_weight):
    if total_weight > 50:
        return total_weight * 7
    else:
        return total_weight * 10

        
df['total_weight'] = df['unit_weight'] * df['qty']
df['shipping_price'] = df['total_weight'].apply(shipping_calculate)
df.head()




#question: it seems as if each row of data is isolated by the specific item purchased, and 
# that a customer who placed a multi-item order will be entered in the database as 1 entry per item contained in the order
# if this is correct, then calculating the shipping price should be done at the ORDER level, not the unit level
# and this particular calculation could possibly result in overcalculating shipping costs PER ORDER. Is that true?
```

```python
# Create a column for the total price using the subtotal and the shipping price along with a sales tax of 9.25%

df['total_price'] = (df['subtotal'] + df['shipping_price']) * 1.095
df.head()


df.head()
```

```python
# Create a column for the cost of each line using unit cost, qty, and
# shipping price (assume the shipping cost is exactly what is charged to the client).
df['line_cost'] = (df['unit_cost'] * df['qty']) + df['shipping_price']
df.head()



```

```python
# Create a column for the profit of each line using line cost and line price
df['profit'] = (df['total_price'] - df['line_cost'])
df.head()
```

## Part 3: Confirm your work
You have email receipts showing that the total prices for 3 orders. Confirm that your calculations match the receipts. Remember, each order has multiple lines.

Order ID 2742071 had a total price of \$152,811.89

Order ID 2173913 had a total price of \$162,388.71

Order ID 6128929 had a total price of \$923,441.25


```python
# Check your work using the totals above

# List of specific order IDs to calculate the total price for
order_ids = [2742071, 2173913, 6128929]

# Filter the DataFrame for only the selected order IDs and make a copy
selected_orders = df[df['order_id'].isin(order_ids)].copy()

# Sum the 'subtotal' and 'shipping_price' per order_id to get the total for each order
order_totals = selected_orders.groupby('order_id').agg({
    'subtotal': 'sum',
    'shipping_price': 'sum'
}).reset_index()

# Calculate the total price for each of these orders using the formula
order_totals['total_price'] = (order_totals['subtotal'] + order_totals['shipping_price']) * 1.0925

# Display the order IDs and their corresponding total prices
print(order_totals[['order_id', 'total_price']])

# Verify with the receipts
expected_prices = {
    2742071: 152811.89,
    2173913: 162388.71,
    6128929: 923441.25
}

# Compare the calculated total prices with the expected values
for order_id in order_ids:
    calculated = order_totals[order_totals['order_id'] == order_id]['total_price'].values[0]
    expected = expected_prices[order_id]
    print(f"Order ID {order_id}:")
    print(f"Calculated Total Price: ${calculated:.2f}")
    print(f"Expected Total Price: ${expected:.2f}")
    print(f"Difference: ${calculated - expected:.2f}\n")
```

## Part 4: Summarize and Analyze
Use the new columns with confirmed values to find the following information.

```python
# How much did each of the top 5 clients by quantity spend? Check your work from Part 1 for client ids.

# find top 5 clients
top_client_data = df[df['client_id'].isin(top_5_clients)]

# calculate total spend for each of the top 5
top_5_total_spend = top_client_data.groupby('client_id')['total_price'].sum().reset_index()

# print the total spending by each of the top 5 clients
print(top_5_total_spend[['client_id', 'total_price']])
```

```python
# Create a summary DataFrame showing the totals for the for the top 5 clients with the following information:
# total units purchased, total shipping price, total revenue, and total profit.

# Step 1: Filter the DataFrame for only the top 5 clients
top_clients_data = df[df['client_id'].isin(top_5_clients)]

# Step 2: Calculate the required totals for each client
summary = top_clients_data.groupby('client_id').agg(
    total_units_purchased=('qty', 'sum'),
    total_shipping_price=('shipping_price', 'sum'),
    total_revenue=('total_price', 'sum'),
    total_profit=('profit', 'sum')
).reset_index()


# Step 3: Print the summary DataFrame
print(summary)

# Format the data and rename the columns to names suitable for presentation.
# Define the money columns.
money_columns =['total_shipping_price', 'total_revenue', 'total_profit']

# Define a function that converts a dollar amount to millions.
def currency_format_millions(x):
    return f"${x / 1_000_000:.2f}M" if pd.notnull(x) else "$0.00M"

# Apply the currency_format_millions function to only the money columns.
summary[money_columns] = summary[money_columns].applymap(currency_format_millions)

# Rename the columns to reflect the change in the money format.
summary.rename(columns={
               'total_units_purchased': 'Total Units Purchased',
               'total_shipping_price': 'Total Shipping Price (in millions)',
               'total_revenue': 'Total Revenue (in millions)',
               'total_profit': 'Total Profit (in millions)'
               }, inplace=True)
print(summary)
```

```python
# Sort the updated data by "Total Profit (millions)" form highest to lowest and assign the sort to a new DatFrame.

sorted_summary = summary.sort_values(by='Total Profit (in millions)', ascending=False)
print(sorted_summary)
```
